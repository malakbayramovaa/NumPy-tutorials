```python
low_var = np.array([1,1,1,1,1,1,2,2])
high_var = np.array([100,200,300,400,500,600,700,800])

np.std(low_var), np.std(high_var), np.var(low_var), np.var(high_var) # std -> standard variation
```

```python
np.argmax(high_var)    # en boyuk reqemin indeksini qaytarir.
```

```python
np.max(high_var)    # en boyuk reqemi qaytarir
```

```python
np.argmin(high_var )    # en kicik reqemin indeksini qaytarir
```
